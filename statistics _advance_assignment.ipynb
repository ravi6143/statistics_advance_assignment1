There are several types of probability distributions in statistics, each describing how the probabilities of different outcomes are distributed for a particular random variable or process. The choice of distribution depends on the nature of the data and the underlying processes being modeled. Here are some common types of probability distributions:

(1). Uniform Distribution: In a uniform distribution, all possible outcomes are equally likely. It is often used when you have no reason to believe that one outcome is more probable than another within a specified range.

(2). Normal Distribution (Gaussian Distribution): The normal distribution is one of the most important and widely used probability distributions. It has a bell-shaped curve and is characterized by its mean (average) and standard deviation. Many real-world phenomena, such as heights and IQ scores, follow a normal distribution.

(3). Binomial Distribution: The binomial distribution models the number of successes in a fixed number of independent Bernoulli trials (experiments with two possible outcomes, often called "success" and "failure"). It is used for situations like coin flips or counting the number of successes in a series of independent yes/no experiments.

(4). Poisson Distribution: The Poisson distribution is used to model the number of events occurring in a fixed interval of time or space, given a known average rate of occurrence. It's commonly used for counting rare events, such as accidents, phone calls, or arrivals at a service center.

(5). Lognormal Distribution: The lognormal distribution arises when the natural logarithm of a random variable follows a normal distribution. It's often used for modeling data that is inherently positive and right-skewed, such as financial returns or the size of biological organisms.

(6). Chi-Square Distribution: The chi-square distribution is used in hypothesis testing and confidence interval construction. It arises in various statistical tests, such as the chi-square test for independence or the analysis of variance (ANOVA).

```python
import scipy.stats as stats
import matplotlib.pyplot as plt

import numpy as np

def normal_pdf(mean, std):
    
    x = np.linspace(mean -3 * std , mean +3 * std , 200)
    
    pdf_values = stats.norm.pdf(x, loc = mean , scale = std)
    
    
    plt.plot(x ,pdf_values , label = (f'Mean ={mean}', f'Std = {std}'))
    
    plt.title("Normal Distribution ")
    
    plt.xlabel('x')
    
    plt.ylabel('Pdf')
    
    plt.legend()
    
    plt.grid(True)
    
    plt.show()
    

    
    
mean1 = 2
std1 = 1.5
normal_pdf(mean1,std1)



mean2 = 5
std2 = 3

normal_pdf(mean2, std2)

```

The Binomial distribution is a discrete probability distribution that models the number of successes (usually denoted as "k") in a fixed number of independent Bernoulli trials (experiments with two possible outcomes: success and failure). 

The Binomial distribution has several key properties:

(1) .Fixed Number of Trials (n): The distribution is defined for a fixed number of trials or experiments, denoted as "n."

(2). Two Possible Outcomes: Each trial has only two possible outcomes: success (often denoted as "S") and failure (often denoted as "F").

(3). Constant Probability of Success (p): The probability of success (P(S)) remains constant from trial to trial. This probability is denoted as "p."

(4). Independence: The trials are assumed to be independent, meaning the outcome of one trial does not affect the outcome of any other trial.

(5). Discreteness: The Binomial distribution is a discrete probability distribution, meaning that the random variable representing the number of successes can only take on integer values.





Examples of events where the Binomial distribution can be applied:

1 .Coin Flips: When you flip a fair coin (a coin with a 50% chance of heads and 50% chance of tails) multiple times, the number of heads in a fixed number of flips follows a Binomial distribution. For example, you can use the Binomial distribution to calculate the probability of getting exactly 3 heads in 5 coin flips.

2 .Quality Control: In a manufacturing process, you might have a product with a known defect rate (probability of being defective). The number of defective products in a batch of fixed size (e.g., 100) can be modeled using a Binomial distribution. For instance, you can use it to calculate the probability of finding fewer than 10 defective items in a batch with a known defect rate.

(a). Nature of Events:

*.Binomial Distribution: The Binomial distribution is used to model the number of successes (usually denoted as "k") in a fixed number of independent Bernoulli trials (experiments with two possible outcomes: success and failure). Each trial has a constant probability of success.

*.Poisson Distribution: The Poisson distribution is used to model the number of rare events occurring in a fixed interval of time or space. It is suitable for situations where events happen randomly and independently, but the average rate of occurrence is known.


(b). Number of Trials:

*.Binomial Distribution: The Binomial distribution requires a fixed number of trials (denoted as "n") to be specified in advance.

*.Poisson Distribution: The Poisson distribution does not require the number of trials to be fixed in advance; it models events occurring over a continuous interval.


(c). Probability of Success:

*.Binomial Distribution: In the Binomial distribution, each trial has a constant probability of success (denoted as "p"). This probability remains the same from trial to trial.

*.Poisson Distribution: The Poisson distribution does not involve a fixed probability of success for individual trials. Instead, it is characterized by the average rate of event occurrence (denoted as "").


(d). Distribution Parameters:

*.Binomial Distribution: The Binomial distribution is parameterized by two values: the number of trials (n) and the probability of success (p).

*.Poisson Distribution: The Poisson distribution is parameterized by a single value: the average rate of event occurrence ().


(e). Use Cases:

*.Binomial Distribution: It is commonly used when you have a fixed number of trials and you want to calculate the probability of a specific number of successes in those trials. *.Poisson Distribution: It is used when you want to model rare events happening in a continuous interval or space.
